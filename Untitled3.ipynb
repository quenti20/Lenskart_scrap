```python
import pandas as pd

df = pd.read_csv('product_detail 1.csv')
products_id = df['id'].tolist()
len(products_id) 
print(products_id[0])
```

```python
import requests
```

```python
for product_id in products_id:
    for i in range():
        url = f'https://api-gateway.juno.lenskart.com/v2/products/product/{product_id}/review?count=1000&page={i}'
        res = requests.get(url)
        print(res.status_code)
```

```python
dfs = []
for product_id in products_id[:2]:
    product_dfs= []
    url = f'https://api-gateway.juno.lenskart.com/v2/products/product/{product_id}/review?count=1000&page=1'
    res = requests.get(url)
    print(res.status_code)
    data = res.json()
    reviews = data["result"]["review"]["reviews"]
    df = pd.DataFrame(reviews)
    df['product_id'] = product_id
    product_dfs.append(df)
product_df = pd.concat(product_dfs, ignore_index=True)
dfs.append(product_df)



#df['product_id'].unique()
```

```python
import requests
import pandas as pd

dfs = []

# Initialize product_dfs outside the loop
product_dfs = []
count = 0
for product_id in products_id:
    url = f'https://api-gateway.juno.lenskart.com/v2/products/product/{product_id}/review?count=1000&page=1'
    res = requests.get(url)
    print(res.status_code)
    count += 1
    print(count)
    print(product_id)
    data = res.json()
    reviews = data["result"]["review"]["reviews"]
    
    df = pd.DataFrame(reviews)
    df['product_id'] = product_id
    
    product_dfs.append(df)

product_df = pd.concat(product_dfs, ignore_index=True)

dfs.append(product_df)

all_reviews_df = pd.concat(dfs, ignore_index=True)

all_reviews_df.to_csv('reviews.csv',index=False)
```
