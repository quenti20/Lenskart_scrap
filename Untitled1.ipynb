```python
import requests
import pandas as pd
```

```python
category_id="3363"
page_size="1400"

dfs = []

for i in range(4):

    res = requests.get('https://api-gateway.juno.lenskart.com/v2/products/category/' + category_id + '?page-size=' + page_size+"&page="+str(i))

    data = res.json()

    with open(category_id+".json", "wb") as file:

        file.write(res.content)

    products = []

    for result in data["result"]["product_list"]:

        product_info = result.copy()


        market_price = None

        lenskart_price = None

        currency_code = None

        frame_size_id = None

        frame_size_name = None

        frame_shape_id = None

        frame_shape_name = None

        brandname_id = None

        brandname_name = None

        frametype_id = None

        frametype_name = None

        for price_info in result["prices"]:

            if price_info["name"] == "Market Price":

                market_price = price_info["price"]

                currency_code = price_info["currency_code"]

            elif price_info["name"] == "Lenskart Price":

                lenskart_price = price_info["price"]

        for tag in result["hashtagList"]:

            if tag["property"] == "frame_size_id":

                frame_size_id = tag["id"]

                if "name" in tag:

                    frame_size_name = tag["name"]

            elif tag["property"] == "frame_shape_id":

                frame_shape_id = tag["id"]

                if "name" in tag:

                    frame_shape_name = tag["name"]

            elif tag["property"] == "brandname_id":

                brandname_id = tag["id"]

                if "name" in tag:

                    brandname_name = tag["name"]

            elif tag["property"] == "frametype_id":

                frametype_id = tag["id"]

                if "name" in tag:

                    frametype_name = tag["name"]

        product_info.update({

            "market_price": market_price,

            "lenskart_price": lenskart_price,

            "currency_code": currency_code,

            "frame_size_id": frame_size_id,

            "frame_size_name": frame_size_name,

            "frame_shape_id": frame_shape_id,

            "frame_shape_name": frame_shape_name,

            "brandname_id": brandname_id,

            "brandname_name": brandname_name,

            "frametype_id": frametype_id,

            "frametype_name": frametype_name,

        })

        products.append(product_info)

    df = pd.DataFrame(products)

    dfs.append(df)

result_df = pd.concat(dfs)

result_df.to_csv('product_detail.csv', index=False)
```
